· 以下采用填充均值方式：

```python
df_1['Age'].fillna(df_1['Age'].mean())
```

### 其他DF操作

#### 删除重复值

根据Name删除重复值，保留第一个出现的记录：df_1.drop_duplicates('Name', keep='first', inplace=False)--inplace指是直接在原来数据上修改还是保留一个副本  


#### 数据替换

· df_1.replace('female','1')

#### 数据表合并

· 上下合并：
df_merged = pd.concat([df1,df2,df3],axis=0,ignore_index=True)  
· 参数axis=0表示上下合并，1表示左右合并，ignore_index=True表示忽略原来的索引

· merge：  
· pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,left_index=False, right_index=False)  
· how代表方式，inner、outer、left、right与SQL中原理相似   
· on用于左右表用于join的主键有一样的列名。若列名不同则分别规定lefton和righton

#### 索引

· 行（列）选取（单维度选取）：df[]。这种情况一次只能选取行或者列，即一次选取中，只能为行或者列设置筛选条件（只能为一个维度设置筛选条件）。   
· 区域选取（多维选取）：df.loc[]，df.iloc[]

```python
#只选取列
#df_1['Name']
df_1[['Name','Pclass']]
```

```python
#只选取行
df_1[1:4]
```

```python
#通过条件选取
df_1[(df_1['Age']>70)]
```

```python
df_1.loc[8]#根据行索引
```

```python
df_1.iloc[1:5,:]#根据行和列索引[行的范围，列的范围]
```

```python
df_1.loc[(df_1['Age']>70)]
```

```python
df_1.loc[(df_1['Age']>25)&(df_1['Pclass']==2),['Name','Sex','Age','Fare','Pclass']]
```

## 统计神器：pivot table

pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, dropna=True, margins_name='All')  
· values:进行透视的目标值  
· index:行   
· columns：列  
· aggfun：进行聚合的函数，如count、mean、sum  
· margins:是否需要加总

```python
df_pivot = pd.pivot_table(df_1,values = 'Age',index = 'Sex',columns = 'Pclass',aggfunc='max',margins = True)
```
