### 第一步：在本地建立数据库

```python
import sqlite3
from sqlite3 import Error
 
 
def create_connection(db_file):
    """ create a database connection to a SQLite database """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        if conn:
            conn.close()
 
 
if __name__ == '__main__':
    create_connection("testnew.db")
```

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('testnew.db')

df = pd.read_csv("test.csv")#把csv导入成为一张数据表

df.to_sql("test", conn, if_exists='append', index=False)
```

### 第三步：SQL实操

```python
import sqlite3

conn = sqlite3.connect("testnew.db")
cur = conn.cursor()
cur.execute("SELECT PassengerId,Pclass,Name,Age FROM test WHERE Name LIKE '%kate%' AND ....")#执行sql语句

rows = cur.fetchall()

pd.DataFrame(rows,)
```

```python
cur.execute("SELECT Pclass,COUNT(DISTINCT PassengerId) FROM test GROUP BY Pclass")#执行sql语句

rows = cur.fetchall()
#rows
pd.DataFrame(rows,)
```

```python
cur.execute("SELECT DISTINCT a.PassengerId,a.Money,b.Pclass,b.Name,b.age FROM test1 a LEFT JOIN (SELECT PassengerId,Pclass,Name,Age FROM test WHERE Age > 30) b on a.PassengerId = b.PassengerId")#执行sql语句

rows = cur.fetchall()
#rows
pd.DataFrame(rows,)
```

```python
cur.execute('SELECT t1.Pclass,t1.COUNT_P FROM( SELECT Pclass,count(PassengerId)AS COUNT_P FROM test GROUP BY Pclass)t1 WHERE COUNT_P >100')

rows = cur.fetchall()
#rows
pd.DataFrame(rows,)
```

```python
cur.execute(' SELECT PassengerId, (CASE WHEN Money > 500 THEN "RICH" ELSE "POOR" END) AS Money_Classification FROM test1')

rows = cur.fetchall()
#rows
pd.DataFrame(rows,)
```

```python
cur.execute(' SELECT SUM(CASE WHEN Money > 500 THEN 1 ELSE 0 END) RICHMEN,SUM(CASE WHEN Money <= 500 THEN 1 ELSE 0 END) AS POORMEN FROM test1')

rows = cur.fetchall()
#rows
pd.DataFrame(rows,)
```
